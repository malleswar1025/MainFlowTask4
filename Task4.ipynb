```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('USvideos.csv')
```

```python
df.head()
```

```python
# Display the summary statistics of the dataset
print("\nSummary statistics of the dataset:")
display(df.describe(include='all'))
```

```python
# Checking for missing values
print("\nChecking for missing values:")
display(df.isnull().sum())
```

```python
# Checking the data types of each column
print("\nData types of each column:")
display(df.dtypes)
```

```python
# Selecting only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
```

```python
# Plotting the distribution of numerical variables
print("\nDistribution of numerical variables:")
numeric_df.hist(figsize=(14, 10), bins=30)
plt.show()
```

```python
# Boxplot to identify outliers
print("\nBoxplot for identifying outliers:")
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Checking correlations between variables
print("\nCorrelation matrix (numeric columns only):")
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Pairplot to visualize relationships between variables
print("\nPairplot of the dataset (numeric columns only):")
sns.pairplot(numeric_df)
plt.show()
```
